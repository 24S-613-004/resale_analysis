# Libraries

```python
import os
import pandas as pd
import glob
```

# Compile

```python
file_path = "data/*.csv"
file_list = glob.glob(file_path)
file_list
```

```python
df = pd.DataFrame()

for file in file_list:
    df1 = pd.read_csv(file)
    df = pd.concat([df,df1], ignore_index = True)
    
df
```

```python
df[['brand','category']].value_counts()
```

# Remove Duplicates


```python
[df.columns]
```

```python
df = df.sort_values(by = ['title', 'price', 'size', 'category', 'style','brand','ingest_timestamp'],
                    ascending = [True,True,True,True,True,True,False])

df = df.drop_duplicates(subset = ['title', 'price', 'size', 'category', 'style','brand'], keep = 'first')
```

# Write

```python
df.to_csv("resell_prices.csv",index = False)
```
